```python
numerical_features = ['MSSubClass', 'LotFrontage', 'LotArea', 'OverallQual', 'OverallCond',
       'YearBuilt', 'YearRemodAdd', 'MasVnrArea', 'BsmtFinSF1',
       'BsmtUnfSF', 'TotalBsmtSF', '1stFlrSF', '2ndFlrSF',
       'GrLivArea', 'BsmtFullBath', 'FullBath', 'HalfBath',
       'BedroomAbvGr', 'TotRmsAbvGrd', 'Fireplaces',
       'GarageYrBlt', 'GarageCars', 'GarageArea','MoSold', 'YrSold']
categorical_features = ['MSZoning', 'LotShape', 'LandContour', 'LotConfig',
       'LandSlope', 'Neighborhood', 'Condition1', 'Condition2', 'BldgType',
       'HouseStyle', 'RoofStyle', 'Exterior1st', 'Exterior2nd',
       'MasVnrType', 'ExterQual', 'ExterCond', 'Foundation', 'BsmtQual',
       'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2',
       'HeatingQC', 'Electrical', 'KitchenQual',
       'FireplaceQu', 'GarageType', 'GarageFinish', 'GarageQual', 'GarageCond',
       'SaleType', 'SaleCondition']
```

```python
from sklearn.model_selection import GridSearchCV, cross_val_score
from sklearn.preprocessing import MinMaxScaler, StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import RandomForestRegressor
from catboost import CatBoostRegressor
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
```

```python
def experiment(model, scaler):
    preprocessing=ColumnTransformer(transformers=[
        ('scaling', scaler, numerical_features),
        ('encoding', OneHotEncoder(drop='first', handle_unknown='ignore'), categorical_features)
    ])

    pipeline=Pipeline(steps=[
        ('preprocessing', preprocessing),
        ('modeling', model)
    ])

    pipeline.fit(x, y)
    result = cross_val_score(pipeline, x, y)
    prediction = pipeline.predict(x)
    print('r2 score =', r2_score(y, prediction))
    print('MSE =', mean_squared_error(y, prediction))
    print('MAE =', mean_absolute_error(y, prediction))
    print("""
    *Evaluation for the model performance* :
    --------------------------------------""")
    print(result)
    print('cross val mean =', result.mean()*100)
    print('Standard Deviation =', result.std())
```

```python
experiment(KNeighborsRegressor(), StandardScaler())
```

```python
experiment(CatBoostRegressor(verbose=0), MinMaxScaler())
```

## Hyperparameter Tuning for CatBoostRegressor

```python
preprocessing=ColumnTransformer(transformers=[
    ('scaling', MinMaxScaler(), numerical_features),
    ('encoding', OneHotEncoder(drop='first', handle_unknown='ignore'), categorical_features)
])

pipeline=Pipeline(steps=[
    ('preprocessing', preprocessing),
    ('modeling', CatBoostRegressor(verbose=0))
])

# hyperparameter tuning
param_grid = {
    'modeling__depth': [6, 7],
    'modeling__learning_rate': [0.1, 0.2],
    'modeling__iterations': [500, 600]
}
grid = GridSearchCV(pipeline, param_grid=param_grid)
grid.fit(x, y)
prediction = grid.predict(x)
best_model = grid.best_estimator_
print('MSE =', mean_squared_error(y, prediction))
print('MAE =', mean_absolute_error(y, prediction))
result = cross_val_score(best_model, x, y, cv=5)
print("""
*Evaluation for the model performance* :
--------------------------------------""")
print(result)
print('cross val mean =', result.mean()*100)
print('Standard Deviation =', result.std())
```
